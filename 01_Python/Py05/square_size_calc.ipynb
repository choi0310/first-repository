```python
class Square:
    def __init__(self):
        # 넓이를 계산할 때 사용할 가로, 세로, 높이 등의 변수를 초기화합니다.
        self.width = 0
        self.height = 0
        self.lower = 0  # 사다리꼴의 아랫변
        self.upper = 0  # 사다리꼴의 윗변

    # 1. 사각형 넓이 구하기
    def rect(self):
        self.width = int(input('가로를 입력하세요: '))
        self.height = int(input('세로를 입력하세요: '))
        area = self.width * self.height
        print(f'사각형의 넓이는 {area}입니다.')
        return area

    # 2. 평행사변형 넓이 구하기 (밑변 * 높이)
    def parall(self):
        # self 인자를 활용하여 인스턴스 변수에 값을 할당합니다.
        self.width = int(input('밑변을 입력하세요: '))
        self.height = int(input('높이를 입력하세요: '))
        
        # 평행사변형 넓이 공식: 밑변 * 높이
        area = self.width * self.height
        print(f'평행사변형의 넓이는 {area}입니다.')
        return area

    # 3. 사다리꼴 넓이 구하기 ((윗변 + 아랫변) * 높이 / 2)
    def trape(self):
        # self 인자를 활용하여 사다리꼴 전용 변수를 사용합니다.
        self.upper = int(input('윗변을 입력하세요: '))
        self.lower = int(input('아랫변을 입력하세요: '))
        self.height = int(input('높이를 입력하세요: '))
        
        # 사다리꼴 넓이 공식: (윗변 + 아랫변) * 높이 / 2
        area = (self.upper + self.lower) * self.height / 2
        print(f'사다리꼴의 넓이는 {area}입니다.')
        return area
```
