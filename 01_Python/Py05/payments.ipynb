```python
# 지불 메서드 (앞서 작성한 pay 함수를 클래스 메서드로 통합)
    def pay(self):
        while True:
            payment_method = input("결제 방법을 선택하세요 ( cash = 1 , card = 2 ) : ")
            
            # 현금을 선택한 경우
            if payment_method == 'cash' or payment_method == '1':
                print("직원을 호출하겠습니다.")
                break  # 올바른 입력이므로 반복문 탈출
                
            # 카드를 선택한 경우
            elif payment_method == 'card' or payment_method == '2':
                print("IC칩 방향에 맞게 카드를 꽂아주세요.")
                break  # 올바른 입력이므로 반복문 탈출
                
            # 그 외의 잘못된 입력을 했을 경우
            else:
                print("★ 잘못된 입력입니다. 다시 결제를 시도해 주세요.")

    # 주문서 출력 메서드 (앞서 작성한 table 함수를 클래스 메서드로 통합)
    def table(self):
        # 외곽 상단 출력
        print('⟝' + '-' * 30 + '⟞')
        for i in range(5):
            print('|' + ' ' * 31 + '|')

        # 주문 상품명 : 해당 금액 반복 출력
        # self.order_menu와 self.order_price에 쌓인 데이터를 활용합니다.
        for i in range(len(self.order_menu)):
            print(f"{self.order_menu[i]} : {self.order_price[i]}")

        # sum() 함수를 이용한 합계 금액 출력
        print('합계 금액 :', sum(self.order_price))

        # 외곽 하단 출력
        for i in range(5):
            print('|' + ' ' * 31 + '|')
        print('⟝' + '-' * 30 + '⟞')
```
