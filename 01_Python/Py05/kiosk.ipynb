```python
class Kiosk:
    def __init__(self):
        # 1. 메뉴판 및 가격표 초기화
        self.menu = ['americano', 'latte', 'mocha', 'choco_latte']
        self.price = [2000, 3000, 3000, 3000]
        
        # 2. 주문 내역을 저장할 리스트 초기화
        self.order_menu = []
        self.order_price = []

    # 메뉴 출력 메서드
    def menu_print(self):
        print("\n" + "="*10 + " MENU " + "="*10)
        for i in range(len(self.menu)):
            print(f"{i+1}. {self.menu[i]} : {self.price[i]}원")
        print("="*26)

    # 주문 및 추가 주문 메서드
    def menu_select(self):
        print("\n[ 주문을 시작합니다 ]")
        
        while True:
            # 1. 음료 번호 선택 및 예외 처리
            try:
                n = int(input("음료 번호를 선택하세요 : "))
            except ValueError:
                print("★ 숫자로 입력해 주세요.")
                continue

            # 메뉴판에 있는 번호인지 확인
            if 1 <= n <= len(self.menu):
                # 2. 온도 선택 및 예외 처리 (HOT / ICE)
                while True:
                    t = input("온도를 선택하세요 ( HOT = 1, ICE = 2 ) : ")
                    if t == '1' or t.upper() == 'HOT':
                        temp = 'HOT'
                        break
                    elif t == '2' or t.upper() == 'ICE':
                        temp = 'ICE'
                        break
                    else:
                        print("★ 잘못된 입력입니다. 1 또는 2를 입력해 주세요.")
                
                # 주문 리스트에 추가 (예: HOT americano)
                selected_item = f"{temp} {self.menu[n-1]}"
                selected_price = self.price[n-1]
                
                self.order_menu.append(selected_item)
                self.order_price.append(selected_price)
                print(f"-> [추가 완료] {selected_item} ({selected_price}원)")
                
                # 3. 추가 주문 여부 확인 (루브릭 핵심 요구사항)
                while True:
                    additional = input("추가 주문을 하시겠습니까? ( 예 = 1, 아니오 = 2 ) : ")
                    if additional == '1' or additional == '예':
                        self.menu_print() # 메뉴판 다시 보여주기
                        break # 안쪽 루프를 빠져나가 위쪽 음료 선택으로 이동
                    elif additional == '2' or additional == '아니오':
                        print("\n[ 주문이 완료되었습니다. ]")
                        return # menu_select 메서드 전체를 종료하고 나감
                    else:
                        print("★ 잘못된 입력입니다. 1 또는 2를 입력해 주세요.")

            else:
                print(f"★ 메뉴판에 없는 번호입니다. 1부터 {len(self.menu)} 사이의 숫자를 입력해 주세요.")
```
