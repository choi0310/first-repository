```python
import random


# Account 클래스 생성
class Account:

    # 클래스 변수
    account_count = 0

    # 생성자
    def __init__(self, name, money):

        # 은행 이름
        self.bank_name = "SC은행"

        # 예금주
        self.user = name

        # 잔액
        self.money = money

        # 입금 횟수
        self.deposit_count = 0

        # 입금 내역 리스트
        self.deposit_list = []

        # 출금 내역 리스트
        self.withdraw_list = []

        # 계좌번호 랜덤 생성
        front = random.randint(100, 999)
        middle = random.randint(10, 99)
        back = random.randint(100000, 999999)

        self.account_number = f"{front}-{middle}-{back}"

        # 계좌 개수 증가
        Account.account_count += 1

    # 계좌 개수 확인
    def get_account_num(self):

        print("현재 개설된 계좌 수 :",
              Account.account_count)

    # 입금 함수
    def deposit(self, price):

        # 1원 이상 입금 가능
        if price < 1:

            print("1원 이상 입금 가능합니다")
            return

        # 잔액 증가
        self.money += price

        # 입금 횟수 증가
        self.deposit_count += 1

        # 입금 내역 저장
        self.deposit_list.append(price)

        print(price, "원 입금 완료")
        print("현재 잔액 :", format(self.money, ","), "원")

        # 입금 5회마다 이자 지급
        if self.deposit_count % 5 == 0:

            interest = self.money * 0.01

            self.money += interest

            print("이자 지급 완료 :", format(int(interest), ","), "원")
            print("현재 잔액 :", format(int(self.money), ","), "원")

        print()

    # 출금 함수
    def withdraw(self, price):

        # 잔액 부족 체크
        if price > self.money:

            print("잔액 부족")
            return

        # 출금
        self.money -= price

        # 출금 내역 저장
        self.withdraw_list.append(price)

        print(price, "원 출금 완료")
        print("현재 잔액 :", format(self.money, ","), "원")
        print()

    # 계좌 정보 출력
    def display_info(self):

        print("은행이름 :", self.bank_name)
        print("예금주 :", self.user)
        print("계좌번호 :", self.account_number)
        print("잔고 :", format(int(self.money), ","), "원")

    # 입금 내역 출력
    def deposit_history(self):

        print("===== 입금 내역 =====")

        for i in self.deposit_list:

            print(format(i, ","), "원")

        print()

    # 출금 내역 출력
    def withdraw_history(self):

        print("===== 출금 내역 =====")

        for i in self.withdraw_list:

            print(format(i, ","), "원")

        print()


# 객체 생성
a1 = Account("kim", 1000000)
a2 = Account("lee", 50000)
a3 = Account("park", 3000000)

# 리스트 저장
account_list = [a1, a2, a3]


# 입금
a1.deposit(5000)
a1.deposit(2000)
a1.deposit(3000)
a1.deposit(4000)
a1.deposit(5000)

# 출금
a1.withdraw(10000)

# 계좌 정보 출력
a1.display_info()

print()

# 입금 내역 출력
a1.deposit_history()

# 출금 내역 출력
a1.withdraw_history()

# 계좌 개수 출력
a1.get_account_num()

print()

# 잔액 100만원 이상 고객 출력
print("===== 100만원 이상 고객 =====")

for i in account_list:

    if i.money >= 1000000:

        i.display_info()
        print()
```
