```python
# 랜덤 모듈 import
import random


# Character 클래스 생성
class Character:

    # 생성자
    def __init__(self, name, level, hp, attack, defence):

        #   캐릭터 이름
        self.name = name

        #   캐릭터 레벨
        self.level = level

        #   캐릭터 체력
        self.hp = hp

        #   캐릭터 공격력
        self.attack = attack

        #   캐릭터 방어력
        self.defence = defence

    # 살아있는지 확인
    def is_alive(self):

        if self.hp > 0:
            return True

        else:
            return False

    # 데미지 받기
    def take_damage(self, damage):

        # 방어력보다 공격력이 높을 때만 피해
        if damage > self.defence:

            #   입은 데미지 만큼 캐릭터 체력 감소
            self.hp -= (damage - self.defence)

        # 체력이 0 밑으로 내려가지 않게
        if self.hp < 0:
            self.hp = 0
            print(self.name , "님이 사망했습니다")

    # 공격하기
    def attack_target(self, target):

        # 1 ~ self.attack 사이 랜덤 데미지
        damage = random.randint(1, self.attack)

        print(self.name, "의 공격!")
        print("공격 데미지 :", damage)

        # 상대에게 데미지 전달
        target.take_damage(damage)

        print(self.name, "이(가)",
              target.name, "을 공격했습니다")


# Player 클래스 생성
class Player(Character):

    # Character 상속
    def __init__(self, name):

        # 부모 클래스 생성자 호출
        super().__init__(name, 1, 100, 25, 5)

        # 경험치 추가
        self.exp = 0

    # 경험치 획득
    def gain_experience(self, exp):

        self.exp += exp

        print(self.name, "경험치",
              exp, "획득")

    # 레벨업
    def level_up(self):

        # 경험치 50 이상이면
        if self.exp >= 50:

            self.level += 1
            self.attack += 10
            self.defence += 5

            # 경험치 차감
            self.exp -= 50

            print("레벨업!")
            print("현재 레벨 :", self.level)
            print("공격력 :", self.attack)
            print("방어력 :", self.defence)


# Monster 클래스 생성
class Monster(Character):

    # Character 상속
    def __init__(self, name, level):

        # 몬스터 능력치 랜덤 생성
        hp = random.randint(10, 30) * level

        attack = random.randint(5, 20) * level

        defence = random.randint(1, 5) * level

        # 부모 클래스 생성자 호출
        super().__init__(
            name,
            level,
            hp,
            attack,
            defence
        )


# 플레이어 생성
player = Player("용사")

# 몬스터 생성
monster = Monster("슬라임", 2)

# 공격
player.attack_target(monster)

print("몬스터 현재 체력 :", monster.hp)

# 경험치 획득
player.gain_experience(50)

# 레벨업
player.level_up()



# 전투 함수
def battle(player, monster):

    # 둘 중 하나가 죽을 때까지 반복
    while player.is_alive() and monster.is_alive():

        # 플레이어 공격
        player.attack_target(monster)

        print(monster.name, "체력 :", monster.hp)

        # 몬스터가 죽었는지 확인
        if not monster.is_alive():
            break

        # 몬스터 공격
        monster.attack_target(player)

        print(player.name, "체력 :", player.hp)

        print("----------------")

    # 플레이어 승리
    if player.is_alive():

        exp = monster.level * 20

        player.gain_experience(exp)

        player.level_up()

        print("전투 승리!")

    # 플레이어 패배
    else:

        print("전투 패배..")

# 플레이어 생성
player = Player("용사")

# 몬스터 생성
monster = Monster("슬라임", 2)

# 전투 시작
battle(player, monster)


# 메인 함수
def main():

    # 몬스터 딕셔너리
    monster_dict = {
        "슬라임": 1,
        "고블린": 2,
        "오크": 3
    }

    # 사용자 이름 입력
    player_name = "모두연"

    # 플레이어 생성
    player = Player(player_name)

    print("게임 시작!")
    print("----------------")

    # 몬스터 순서대로 전투
    for monster_name, monster_level in monster_dict.items():

        print()
        print(monster_name, "등장!")
        print("레벨 :", monster_level)

        # 몬스터 생성
        monster = Monster(monster_name, monster_level)

        # 전투
        battle(player, monster)

        print("----------------")

        # 플레이어 사망 확인
        if not player.is_alive():

            print("게임 오버")
            return

    print("모든 몬스터 처치 완료!")

main()
```
